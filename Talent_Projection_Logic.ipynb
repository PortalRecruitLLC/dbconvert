```
import numpy as np
from typing import Dict, Any

def compute_nba_translatability(stats: Dict[str, Any], position_group: str) -> float:
    """
    Archon-grade translatability projection.
    Calculates the 'Dog Index' by blending physical disruptor metrics
    with high-volume efficiency.
    """
    ppg_norm = stats.get("ppg_norm", 0.5)
    rpg_norm = stats.get("rpg_norm", 0.5)
    apg_norm = stats.get("apg_norm", 0.5)
    stocks_pg = stats.get("stocks_pg", 0.0) # Steals + Blocks

    # Weighting varies by role—don't treat a PG like a Rim Protector
    weights = {
        "guard": {"ppg": 0.45, "apg": 0.35, "stocks": 0.20},
        "forward": {"ppg": 0.40, "rpg": 0.30, "apg": 0.15, "stocks": 0.15},
        "big": {"ppg": 0.30, "rpg": 0.50, "stocks": 0.20}
    }

    w = weights.get(position_group, weights["forward"])
    raw_score = (w.get("ppg", 0) * ppg_norm +
                 w.get("rpg", 0) * rpg_norm +
                 w.get("apg", 0) * apg_norm +
                 w.get("stocks", 0) * (stocks_pg / 5.0))

    # Triple bonus for high-impact floor (the 'triple-double' logic)
    if stats.get("ppg", 0) >= 12 and stats.get("rpg", 0) >= 5 and stats.get("apg", 0) >= 2.5:
        raw_score += 0.10

    return min(100.0, raw_score * 100.0)
```
